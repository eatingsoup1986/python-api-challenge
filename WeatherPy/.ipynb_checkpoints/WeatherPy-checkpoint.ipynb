# Assumption and Notes for Analysis
* Data is a specific point in time for when the Open Weather data was extracted (date of extract noted in report output CSV(cities_to_equator.csv)).
* Does not look at weather data over a period of time.

```python
# Dependencies
import matplotlib.pyplot as plt
import requests
from scipy import stats
import pandas as pd
from config import api_key
import csv
import kdtree
import os
import citipy
from citipy import citipy
import numpy as np
import pprint
```

```python
#setup API link to Open Weather
url = "http://api.openweathermap.org/data/2.5/weather?q="
api_key = "&appid=" + api_key
city = "London"
units = "&units=imperial"
#response = requests.get(url+city+units+api_key).json()

#print(response)
```

```python
#test citipy setup
city = citipy.nearest_city(51.5085, -0.1257)
city.city_name
```

```python
#Start generating list of cities
# List for holding latitude and longtitude and cities
#set the latitude and longitude of the equatorial region: 
#https://www.nationalgeographic.org/encyclopedia/latitude/#:~:text=The%20Equator%20is%20the%20line,Pole%20is%2090%20degrees%20S.
#The Equator is the line of 0 degrees latitude. Each parallel measures one degree north or south of the Equator, 
#with 90 degrees north of the Equator and 90 degrees south of the Equator. 
#The latitude of the North Pole is 90 degrees N, and the latitude of the South Pole is 90 degrees S.

lat_range = (-90, 90)
lon_range = (-180, 180)

lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lon_range[0], lon_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each latitude, longtitude combination using citipy
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#combine the API call to OpenWeather with the city list generated.


#create dictionary for where data is to sit fromt he api call
city_data = {"Latitude": [],
            "Longtitude": [],
            "City": [],
            "Temp": [],
            "Humidity": [],
            "Cloudiness": [],
            "Wind Speed": [],
            "Date": [],
            "Country": []}

for city in cities:
    try:
        response = requests.get(url+city+units+api_key).json()
        city_data["Latitude"].append(response["coord"]["lat"])
        city_data["Longtitude"].append(response["coord"]["lon"])
        city_data["City"].append(response["name"])
        city_data["Temp"].append(response["main"]["temp"])
        city_data["Humidity"].append(response["main"]["humidity"])
        city_data["Cloudiness"].append(response["clouds"]["all"])
        city_data["Wind Speed"].append(response["wind"]["speed"])
        city_data["Date"].append(response["dt"])
        city_data["Country"].append(response["sys"]["country"])
        print(f"City found: {city}")
    except:
        print(f"City not found:{city}")
    
```

```python
#create data frame from collected information
city_df = pd.DataFrame(city_data)

#fix formatting for date data by converting unix to seconds
#(https://stackoverflow.com/questions/19231871/convert-unix-time-to-readable-date-in-pandas-dataframe)
city_df["Date"] = pd.to_datetime(city_df["Date"], unit="s")

#export to CSV in case API link stops working
city_df.to_csv('cities_to_equator.csv')
city_df.head()
```

```python
#Generate scatter plots (Temperature vs Latitude)
plt.scatter(city_df["Latitude"],city_df["Temp"])
plt.grid()
plt.title("Temperature (C) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (C)")
```

The higher temperature are shown to be around between -20 to +20 degree latitudes. As the cities go below the equator into the negative latitudes, the temperature begins to decrease as well.

# Northern and Southern Hemisphere Analysis

```python
# Seperate between Northern and Southern Hemisphere
northern = city_df.loc[city_df["Latitude"]>0]
southern = city_df.loc[city_df["Latitude"]<0]

#print(northern.count())
#print(southern.count()) - used print to add totals (389 + 169 = 588), same as total city_df as a check

```

## Northern/Southern Hemisphere - Temperature vs Latitude Analysis

```python
# Temperatures vs Latutide (Northern Hemisphere)
# Set the linear regression line
slope, intercept, rvalue, pvalue, stderr = stats.linregress(northern["Latitude"], northern["Temp"])
#Set the linear regression formula
temp_regress = northern["Latitude"] * slope + intercept

plt.scatter(northern["Latitude"],northern["Temp"])
plt.plot(northern["Latitude"],temp_regress,"r", label='y={:.2f}x+{:.2f}'.format(slope,intercept))
plt.legend()
plt.grid()
plt.title("Northern Hemisphere Temperature (C) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (C)")
print(f"Pearson r-value = {rvalue}")
plt.savefig("Temperature_Latitude(Northern)")
```

```python
# Temperatures vs Latutide (Southern Hemisphere)
# Set the linear regression line
slope, intercept, rvalue, pvalue, stderr = stats.linregress(southern["Latitude"], southern["Temp"])
#Set the linear regression formula
temp_regress = southern["Latitude"] * slope + intercept

plt.scatter(southern["Latitude"],southern["Temp"])
plt.plot(southern["Latitude"],temp_regress,"r", label='y={:.2f}x+{:.2f}'.format(slope,intercept))
plt.legend()
plt.grid()
plt.title("Southern Hemisphere Temperature (C) vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature (C)")
print(f"Pearson r-value = {rvalue}")
plt.savefig("Temperature_Latitude(Southern)")
```

### Observations
* A strong correlation was identified between temperature and distance to the equator. 
* Relationship was observed to be more prominent for the Northern Hemispheres (r-value = -0.8779950662859505).
* Southern Hemisphere still had an observed medium correlation for temperature to latitude.
