```python
# Dependencies
import matplotlib.pyplot as plt
import requests
from scipy import stats
import pandas as pd
from config import api_key
import gmaps
#access gmaps using the api key
gmaps.configure(api_key)
```

```python
#import weather data from WeatherPy exported csv file
weather = pd.read_csv("cities_to_equator.csv")
```

```python
weather.head()
```

```python
figure_layout = {
    'width': '900px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
fig
```

```python
#Generate the heat map: https://jupyter-gmaps.readthedocs.io/en/latest/tutorial.html#heatmaps
#https://developers.google.com/maps/documentation/places/web-service/details

locations = weather[["Latitude", "Longtitude"]].astype(float)
humidity = weather["Humidity"].astype(float)

heatmap_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=max(weather["Humidity"]),
                                 point_radius = 2)
fig.add_layer(heatmap_layer)
fig
```

```python
#narrow down to ideal locations
weather_ideal = weather[weather['Temp'].between(20, 35)]
weather_ideal2 = weather_ideal[weather_ideal["Humidity"]<50]
weather_ideal2 = weather_ideal2.dropna()

weather_ideal2.head(100)
```

```python
hotel_df = weather_ideal2
# params dictionary to update each iteration
params = {
    "radius": 5000,
    "types": "lodging",
    "key": api_key
}


for index, row in hotel_df.iterrows():
    # get lat, lng from df
    lat = row["Latitude"]
    lng = row["Longtitude"]

    # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {row['City']}.")
    response = requests.get(base_url, params=params).json()
    # print(json.dumps(response, indent=4, sort_keys=True))
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel is {results[0]['name']}.")
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

```python
hotel_df
```

```python
# Assign the marker layer
locations = hotel_df[["Latitude", "Longtitude"]]
markers = gmaps.marker_layer(locations)

# Using the template add the hotel marks to the heatmap
#https://jupyter-gmaps.readthedocs.io/en/latest/tutorial.html#heatmaps
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""

# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

# Create hotel symbol layer
hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=hotel_info
)
# Add the layer to the map
fig.add_layer(markers)
fig.add_layer(hotel_layer)
fig
```
